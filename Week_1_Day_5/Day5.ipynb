```python
# Dictionary
student = {
    'name': 'Junaid',
    'age': 32,
    'course': 'AI'
}

print(student)

# get
print(student.get('name'))

# update-> it can update multiple values of kkeys at same time
student.update({'name': 'Awais', 'age': 25})
print(student)

# change single item of dict
student['age'] = 33
print(student)

# loop to acces the dictionary
for key, value in student.items():
  print(f'key => {key}, value => {value}')
```

```python
days = {
    1: 'Monday',
    2: 'Tuesday',
    3: 'Wednesday',
    4: 'Thursday',
    5: 'Friday',
    6: 'Saturday',
    7: 'Sunday'
}

# 1. Use try-except to safely handle non-numeric text input
try:
    take_input = int(input('Please write your day number (1-7): '))

    # 2. Check if the number is inside your dictionary keys
    if take_input in days:
        print(take_input)
        print(f'Your day is {days[take_input]}')
    else:
        print('Error: Please enter a number between 1 and 7 only.')

except ValueError:
    print('Error: Please write a valid number only.')
```

```python
# Even numbers
num = 2
while num <= 10:
  print(num)
  num+=2

# Odd numbers
num = 2
while num <= 10:
  if num%2==1:
    print(num)
  num+=1
```

```python
# Break statement
for n in [1,2,3,4,5]:
  if n == 3:
    break
  print(n)

```

```python
# Continue statement
for n in [1,2,3,4,5]:
  if n == 3:
    continue
  print(n)
```

```python
# Pass statement
for n in [1,2,3,4,5]:
  if n == 3:
    pass
  print(n)
```

```python
# Enumrate
students = ['Ali', 'Sara', 'John']
for index, name in enumerate(students):
  print(f"{index+1}: {name}")
```

```python
students = [{'name': 'Ali'}, {'name': 'Sara'}, {'name': 'John'}]
for index, name in enumerate(students, start=1):
  print(f"{index}: {name['name']}")
```
