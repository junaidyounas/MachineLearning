```python
# Celcius temp function

def celsius(temp):
  return (temp -32) * 5/9

celcius_lambda = lambda temp: (temp-32) * 5/9

print(celsius(90))
print(celcius_lambda(90))
```

```python

# decorator
def func_decorator(func):
  def wrapper():
    print('--------- Start --------')
    func()
    print('--------- End ---------')
  return wrapper()

@func_decorator
def pname(name='Junaid'):
  print(name)
```

```python
name_lambda = lambda name='Junaid': print(f'My name is {name}')
print(name_lambda())
```

```python
name = 'Junaid'

def new():
  name = 'Awais'

new()
print(name)
```

```python
# lambda if else function
check = lambda x: "Even" if x%2==0 else 'Odd'

print(check(8))
```

```python
# map function
numbers = [1,2,3,4,5]
results = list(map(lambda x: x*x, numbers))
print(results)
```

```python
# to uppercase
names = ['Junaid', 'Awais', 'Tayyab']

new_names = list(map(str.lower, names))
print(new_names)
```

```python
# iterable
nums = [10, 20, 30]
it = iter(nums)

print(next(it))
print(next(it))
print(next(it))
print(next(it, "No more items"))
```
