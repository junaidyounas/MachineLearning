```python
# constructor and destructor
class Student:
    def __init__(self):
        self.name = ''
        self.age = 0


s1 = Student()
print(s1.name, s1.age)
```

```python
class Student:
    def __init__(self):
        print('Constructor call')

    def hello(self):
        print("Hello")

s = Student()

s.hello()
```

```python
class Student:
    def __del__(self):
        print('I am destroyer')

    def hello(self):
        print('Hello')


s1 = Student()
print(s1)
```

```python
# Inheritance

class Animal(Dog):
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} eats food")

class Dog():
    def bark(self):
        print(f"{self.name} barks")

d1 = Dog('Buddy')
d1.eat()
d1.bark()
```

```python
# MRO
class A():
    def hello(self):
        print('its a hello')

class B(A):
    def hello(self):
        print('its b hello')

class C(A):
    def hello(self):
        print('its c hello')

class D(B, C):
    pass

c1 = D()
c1.hello()
print(D.mro())
```

```python
# Encapsulation
class A():
    def __secret(self):
        print('this is secret function')

a = A()
a.__secret()
```

```python
# Name mangling
class Car:
    def __init__(self):
        self.__ignition = False
        self.lights = False

    def is_ignation(self):
        print(self.__ignition)


c1 = Car()
c1.is_ignation()
print(c1.__dict__)
print(c1._Car__ignition)
```

```python
# Polymorphism

class BankAccount:
    def calculate_interest(self):
        print('Calculating interest')

class SavingAccount:
     def calculate_interest(self):
        print('Calculating interest')

accounts = [
    BankAccount(),
    SavingAccount()
]

# print(accounts)
for account in accounts:
    account.calculate_interest()
```

```python
class Student:

    def __init__(self, marks):
        self.marks = marks

    def __add__(self, other):
        return self.marks - other.marks


s1 = Student(80)
s2 = Student(90)

print(s1 + s2)
```
