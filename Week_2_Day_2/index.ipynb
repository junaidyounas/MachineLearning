```python
# Taking input from user
class student():
    name = ''
    collage = ''
    year = ''

a = student()
a.name = input('Enter your name')
a.collage = input('Enter your collage')
a.year = input('Enter your year')

print(a.name, a.collage, a.year)
```

```python
# constructor and destructor
class Student:
    def __init__(self):
        self.name = ''
        self.age = 0


s1 = Student()
print(s1.name, s1.age)
```

```python
class Student:
    def __init__(self):
        print('Constructor call')

    def hello(self):
        print("Hello")

s = Student()

s.hello()
```

```python
class Student:
    def __del__(self):
        print('I am destroyer')

    def hello(self):
        print('Hello')


s1 = Student()
print(s1)
```

```python
# Inheritance

class Animal(Dog):
    def __init__(self, name):
        self.name = name

    def eat(self):
        print(f"{self.name} eats food")

class Dog():
    def bark(self):
        print(f"{self.name} barks")

d1 = Dog('Buddy')
d1.eat()
d1.bark()
```
