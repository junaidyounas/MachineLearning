```python
#List
items = ['apple', 'banana', 'cherry', '123']
items2 = ['chicken', 'beef']
# sort
items.sort()
print(items)
# pop
items.pop()
print(items)
# index
print(items[1])
# reverse
items.reverse()
print(items)
# insert
items.insert(1, 'orange')
print(items)
# append
items.append('melon')
print(items)
# changing item
items[1] = 'lichy'
print(items)
# remove one
items.remove('lichy')
print(items)
# sort reverse
items.sort(reverse=True)
print(items)
# join two arrays
final = items + items2
print(final)
```
