```python
# Filter in python
a = [1, 2, 3, 4, 5, 6, 7, 8]
new = list(filter(lambda x: x% 2== 0, a))
print(new)
```

```python
# Nested functions
def outer():
    def inner():
        return 'Hello'
    return inner()

print(outer())
```

```python
# map with outer function
a = [1, 2, 3, 4, 5, 6, 7, 8]
def add(x):
    return x+1

new = list(map(add, a))
print(new)
```

```python
# Return the inner function
def multiplier(factor):
    def multiply(x):
        return x*factor
    return multiply

times3 = multiplier(3)
print(times3(4))
```
