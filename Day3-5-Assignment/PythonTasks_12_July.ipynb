```python
"""
Program 1: (If-Else)
Task:
Take a number from the user and check whether it is Even or Odd.
"""
a = int(input('Please enter your number i will let you know its even or odd: '))

if a%2==0:
  print('Its even number')
else:
  print('Its odd number')
```

```python
"""
Program 3: (Nested If)
Task:
Take the student's marks and fee status.
If marks are greater than or equal to 50:
• If fee is paid → Admission Confirmed
• Otherwise → Fee Required
Else → Not Eligible
"""

marks = int(input('Please enter your marks: '))
feeStatus = input('Is your fee is paid: yes/no: ')

if feeStatus.lower() == 'yes':
  if marks >= 50:
    print('Admission Confirmed')
  else:
    print('Fee required')
else:
  print('Not eligible')

```

```python
"""
Program 5: (While Loop)
Task:
Write a Python program to calculate the sum of all numbers from 1 to a given number using
a while loop.
"""

sum = 0
num = int(input('Please enter a number: '))

while num >= 1:
  sum=num+sum
  num-=1

print(f'sum of numbers is: {sum}')



```

```python
"""
Program 11 (Lists)
Task:
Take five student names in a list. Print each name using a for loop.
"""

list1 = []

for i in range(1,6):
  userInput = input(f'{i}: Please enter student name:')
  list1.append(userInput)

for index, name in enumerate(list1):
  print(f'\n{index}: {name.title()}')
```

```python
"""
Program 14: (Sets)
Task:
Create two sets.
Print:
• Union
• Intersection
"""

set1 = {1, 2, 3}
set2 = {1, 4, 5}

# union
union1 = set1 | set2
union2 = set1.union(set2)
print(union1)
print(union2)

intersection1 = set1 & set2
intersection2 = set1.intersection(set2)

print(intersection1)
print(intersection2)
```

```python
"""
Program 17: (List Comprehension)
Task:
Create a list of numbers from 1 to 10. Using list comprehension, create another list
containing the squares of those numbers.
"""

list1 = list(range(1,11))

list2 = [i**i for i in list1]

print(list1)
print(list2)
```

```python
"""
Program 18: (Functions)
Task:
Write a function that prints the multiplication table of a number.
"""
def writeTable(num):
  for i in range(1,11):
    print(f'{num}x{i} = {num*i}')


writeTable(2)
```

```python
"""
Program 19: (Lambda Expression and map())
Task:
Take multiple numbers from the user.
• Use map() to convert them into integers.
• Use a lambda expression to multiply every number by 2.
• Print the new list.
"""
i = 0
numbersList = []
while i < 3:
  inputTake = input('Please enter the number: ')
  numbersList.append(inputTake)
  i+=1

mappedList = list(map(lambda x: int(x)*2, numbersList))
print(mappedList)
```
